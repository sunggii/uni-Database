## Query

### 1. แสดงข้อมูล contries ที่ไม่ซ้ำกันทั้งหมดจากตาราง ```customers```
- ใช้ ```distinct``` เพื่อให้ได้ข้อมูลที่ไม่ซ้ำกัน

````python
%%sql
SELECT distinct country 
FROM customers;
````

### 2. จากตาราง orderdetails ให้แสดง orderNumber, productCode, quanityOrdered, priceEach และจงคำนวนหา total(quanityOrdered * priceEach) และสร้าง colunm เพื่อเก็บผลการคำนวนชื่อ ```lineTotal```

````python
%%sql
SELECT orderNumber, productCode, quantityOrdered, priceEach , 
(quantityOrdered * priceEach) AS lineTotal
FROM orderdetails
Limit 5;
````

### 3. จงใช้ String function ในการแสดงค่าตามนี้
จากตาราง employees
- แสดง colum firstName, lastName
- แสดง colum ชื่อ ```fullName``` โดย colunm นี้จะแสดงชื่อ-สกุล แบบมีช่องระหว่างระหว่างชื่อ-สกุล
- แสดง colum ชื่อ ```emailUsername``` โดยให้แสดงเฉพาะ username เท่านั้น

````python
%%sql
SELECT firstName, lastName, 
concat(firstName, ' ', lastName) AS fullName,
split_part(email,'@',1) AS emailUsername
FROM employees LIMIT 5;
````

### 4. จงใช้ String function ในการแสดงค่าตามนี้
จากตาราง products
- productName ใน lower case
- productName ใน upper case
- แสดง length ของ productName 



````python
%%sql
SELECT lower(productName) AS lowerName,
upper(productName) AS upperName,
length(productName) AS nameLength
FROM products
LIMIT 5;
````

### 5. จากตาราง products จงแสดง productName, buyPrice, msrp, และสร้าง boolean colunm ชื่อ ```isProfitable``` เพื่อเช็คว่า msrp ดีกว่า buyPrice หรือไม่

````python
%%sql
SELECT productName, buyPrice, MSRP,
(MSRP > buyPrice) AS isProfitable
FROM products 
LIMIT 5;
````
