## Query

### 1. หาทุก Product ที่มี ```buyPrice``` มากกว่า 50

````python
%%sql
SELECT productCode, productName, buyPrice
FROM products 
WHERE buyPrice > 50
LIMIT 5;
````

### 2. หาทุก order ที่สั่งเข้ามาระหว่างวันที่ 1/1/2004 - 31/12/2004

````python
%%sql
SELECT orderNumber, orderDate, customerNumber
FROM orders
WHERE orderDate between '2004-01-01' and '2004-12-31'
LIMIT 5;
````

### 3. หาลูกค้าทุกคนที่อยู่ใน USA, France หรือ Germany

````python
%%sql
SELECT customerNumber, customerName, country
FROM customers
WHERE country in ('USA', 'France', 'Germany')
LIMIT 5;
````

### 4. ค้นหา employee ทั้งหมดที่มีคำว่า ```Sales``` อยู่ในชื่อตำแหน่งงาน และ ```office code``` เป็น 1 ไม่ก็ 2

````python
%%sql
SELECT employeeNumber, firstName, lastName, jobTitle, officeCode
FROM employees
WHERE jobTitle LIKE '%Sales%' and officeCode in ('1', '2')
````

### 5. หาทุก Product ทั้งหมดจาก productLine ```Classic Cars``` ที่มีจำนวนสินค้าในสต็อกน้อยกว่า 1000 ชิ้น หรือ ราคาซื้อมากกว่า 75

````python
%%sql
SELECT productCode, productName, productLine, quantityInStock, buyPrice
FROM products
WHERE productLine in ('Classic Cars') and (quantityInStock < 1000 or buyPrice > 75)
````
