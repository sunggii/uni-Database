## Query

### 1. Find all customers who have had at least one payment

```python
%%sql
SELECT 
    customerNumber,
    customerName,
    city
FROM customers 
WHERE customerNumber in (
    SELECT DISTINCT customerNumber
    FROM payments
)
LIMIT 5;
```

### 2. List products that have been ordered

```python
%%sql
SELECT 
    productCode,
    productName,
    productLine
FROM products
WHERE productCode in (
    SELECT DISTINCT productCode
    FROM orderdetails
)
```

### 3. Find employees who work in offices located in the USA

```python
%%sql
SELECT 
    employeeNumber,
    concat(firstName, ' ', lastName) AS empname,
    jobTitle
FROM employees
WHERE officeCode in (
    SELECT officeCode
    FROM offices
    WHERE country = 'USA'
)
```

### 4. Find orders placed by customers from Germany

```python
%%sql
SELECT
    orderNumber,
    orderDate
FROM orders
WHERE customerNumber in (
    SELECT customerNumber
    FROM customers
    WHERE country = 'Germany'
)
```

### 5. Find customers who have never placed an order

```python
%%sql
SELECT
    customerNumber,
    customerName,
    country
FROM customers
WHERE customerNumber not in (
    SELECT DISTINCT customerNumber
    FROM orders
)
```

### 6. List products that have never been ordered

```python
%%sql
SELECT
    productname,
    buyPrice
FROM products
WHERE productCode not in (
    SELECT DISTINCT productCode 
    FROM orderdetails
)
```

### 7. Find employees who are not sales representatives

```python
%%sql
SELECT
    employeeNumber,
    concat(firstName, ' ', lastName) AS empname,
    jobTitle
FROM employees
WHERE jobTitle not in ('Sales Rep', 'Sales Manager')
```

### 8. Find product lines that don’t have any products priced above $100

```python
%%sql
SELECT
    DISTINCT productLine
FROM products
WHERE productLine not in (
    SELECT productLine
    FROM products
    WHERE buyPrice > 100
)
```

### 9. Find products more expensive than ANY product in the 'Classic Cars' product line

```python
%%sql
SELECT
    productName,
    buyPrice
FROM products
WHERE buyPrice > ANY (
    SELECT buyPrice
    FROM products
    WHERE productLine = 'Classic Cars'
)
AND productLine != 'Classic Cars'
```

### 10. Find customers from countries that have ANY office

```python
%%sql
SELECT
    customerName,
    country
FROM customers
WHERE country = ANY (
    SELECT DISTINCT country
    FROM offices
)
```
